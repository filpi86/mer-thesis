Combine IDs to the songs

```python
# unique IDs 
song_ids = set(main_df['anonymized_author_id'])
tweets_ids = set(tweets_df['anonymized_id'])

common_ids = song_ids.intersection(tweets_ids)

print(f"Number of unique song IDs: {len(song_ids)}")
print(f"Number of unique tweet IDs: {len(tweets_ids)}")
print(f"Number of common IDs: {len(common_ids)}")
```

```python
main_df_merged = main_df.merge(
    tweets_df[['anonymized_id', 'description']],
    left_on='anonymized_author_id',
    right_on='anonymized_id',
    how='left'
)
```

```python
main_df_merged
```

```python
main_df_merged = main_df_merged.drop(columns=["created_at", "music_source", "music_url", "anonymized_tweet_id"])

main_df_merged.head()
```

```python
main_df_merged = main_df_merged[main_df_merged["disorder"].isin(["depression", "control"])]

main_df_merged
```

Get the data frame with the VAs

```python
VA_df = pd.read_csv("/content/drive/MyDrive/Thesis docs/sentiment_song_level_wide_enriched.csv")

VA_df
```

```python
final_df = main_df_merged.merge(VA_df, on='title', how='left')

final_df
```

Repetition diagnose

```python
# shape
print('rows:', len(final_df))
print('columns:', final_df.shape[1])

# Fully duplicated rows
full_dups = final_df.duplicated().sum()
print('fully duplicated rows:', full_dups)

if full_dups > 0:
    print('n\Sample full duplicates:')
    display(final_df[final_df.duplicated(keep=False)].sort_values(list(final_df.columns)).head(10))

# Repetition by key columns
print('\nDuplicated titles:', final_df['title'].duplicated().sum())
print('Duplicated anonymized_author_id:', final_df['anonymized_author_id'].duplicated().sum())

# same author + title repeated
pair_dups = final_df.duplicated(subset=['anonymized_author_id', 'title']).sum()
print('Duplicated author + title pairs:', pair_dups)

if pair_dups > 0:
    print('\nTop repeated (author, title) pairs:')
    display(
        final_df
        .groupby(['anonymized_author_id', 'title'], dropna=False)
        .size()
        .reset_index(name='count')
        .query('count > 1')
        .sort_values('count', ascending=False)
        .head(20)
    )
```

```python
title_q = "Just A Feeling"
aid_q = "046fa37d5fb5ad892208c49e"  # optional but recommended

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 2000)
pd.set_option("display.max_colwidth", None)

print("=== main_df ===")
display(main_df[main_df["title"] == title_q])

print("=== main_df_merged ===")
display(main_df_merged[(main_df_merged["title"] == title_q) & (main_df_merged["anonymized_author_id"] == aid_q)])

print("=== VA_df ===")
display(VA_df[VA_df["title"] == title_q])

print("=== final_df ===")
display(final_df[(final_df["title"] == title_q) & (final_df["anonymized_author_id"] == aid_q)])
```

```python
# VA_df has 4330 duplicate titles — deduplicate by averaging numeric VA columns per title
va_num_cols = VA_df.select_dtypes('number').columns.tolist()
VA_df_deduped = VA_df.groupby('title', as_index=False)[va_num_cols].mean()

# Rebuild final_df using the deduplicated VA_df, then keep one row per (author, title)
final_df_fixed = (
    main_df_merged
    .merge(VA_df_deduped, on='title', how='left')
    .drop_duplicates(subset=['anonymized_author_id', 'title'], keep='first')
    .reset_index(drop=True)
)

print("Rows:", len(final_df_fixed))
print("Duplicated (author, title) pairs:", final_df_fixed.duplicated(subset=['anonymized_author_id', 'title']).sum())
```
