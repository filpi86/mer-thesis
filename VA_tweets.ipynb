Combine IDs to the songs

```python
# unique IDs 
song_ids = set(main_df['anonymized_author_id'])
tweets_ids = set(tweets_df['anonymized_id'])

common_ids = song_ids.intersection(tweets_ids)

print(f"Number of unique song IDs: {len(song_ids)}")
print(f"Number of unique tweet IDs: {len(tweets_ids)}")
print(f"Number of common IDs: {len(common_ids)}")
```

```python
main_df_merged = main_df.merge(
    tweets_df[['anonymized_id', 'description']],
    left_on='anonymized_author_id',
    right_on='anonymized_id',
    how='left'
)
```

```python
main_df_merged
```

```python
main_df_merged = main_df_merged.drop(columns=["created_at", "music_source", "music_url", "anonymized_tweet_id"])

main_df_merged.head()
```

```python
main_df_merged = main_df_merged[main_df_merged["disorder"].isin(["depression", "control"])]

main_df_merged
```
